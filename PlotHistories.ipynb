Notebook to plot histories

```python
import pickle

def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
VU_history=load_obj('vanilla_history')
SU_history=load_obj('satellite_history')
PU_history=load_obj('pretrained_history')
```

Loss history

```python
plt.figure()
plt.title('Loss history')
plt.plot(np.arange(len(VU_history['loss']))+1,VU_history['loss'])
plt.plot(np.arange(len(SU_history['loss']))+1,SU_history['loss'])
plt.plot(np.arange(len(PU_history['loss']))+1,PU_history['loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Vanilla Unet', 'Satellite Unet', 'Pretrained Unet'])
plt.savefig('loss_history.eps')
```

Validation accuracy history

```python
plt.figure()
plt.title('Validation accuracy history')
plt.plot(np.arange(len(VU_history['val_accuracy']))+1,VU_history['val_accuracy'])
plt.plot(np.arange(len(SU_history['val_accuracy']))+1,SU_history['val_accuracy'])
plt.plot(np.arange(len(PU_history['val_accuracy']))+1,PU_history['val_accuracy'])
plt.xlabel('Epoch')
plt.ylabel('Validation accuracy')
plt.legend(['Vanilla Unet', 'Satellite Unet', 'Pretrained Unet'])
plt.savefig('valacc_history.eps')
```

```python
plt.figure()
plt.title('Validation weighted mean IoU history')
plt.plot(np.arange(len(VU_history['val_mean_iou_weighted']))+1,VU_history['val_mean_iou_weighted'])
plt.plot(np.arange(len(SU_history['val_mean_iou_weighted']))+1,SU_history['val_mean_iou_weighted'])
plt.plot(np.arange(len(PU_history['val_mean_iou_weighted']))+1,PU_history['val_mean_iou_weighted'])
plt.xlabel('Epoch')
plt.ylabel('Validation weighted mean IoU')
plt.legend(['Vanilla Unet', 'Satellite Unet', 'Pretrained Unet'])
plt.savefig('valweiiou_history.eps')
```
